```python
#importing libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('World University Rankings 2023.csv')    ## To load or read or import dataset 
df
```

```python
df.shape  #Find out number of rows and columns
```

```python
#CHecking the presence of duplicate rows
df.duplicated().sum()
```

```python
#Dropping duplicate values
df=df.drop_duplicates()
```

```python
#Checking the presence of null values
df.isnull().sum()
```

```python
#Dropping the rows from index 2209 as we dont have name of university
df = df.drop(df.index[2209:])
```

```python
#Filling null values of column name location
if df['Location'].isnull().any():
    # Calculate the mode of the column location
    mode = df['Location'].mode()[0]
    
    # Fill null values with the calculated mode of location
    df['Location'].fillna(mode,inplace=True)

df
```

```python
#FIlling missing value pf no of student per staff column
if df['No of student per staff'].isnull().any():
    # Calculate the median of the column of No of student per staff
    median = df['No of student per staff'].median()
    
    # Fill null values with the calculated median of column No of student per staff
    df['No of student per staff'].fillna(median, inplace=True)

```

```python
#Filling missing value of column research score
if df['Research Score'].isnull().any():
    # Calculate the median of the column research score
    median = df['Research Score'].median()
    
    # Fill null values with the calculated median of research score
    df['Research Score'].fillna(median, inplace=True)

```

```python
#Filling missing value of column teaching score
if df['Teaching Score'].isnull().any():
    # Calculate the median of the column teaching score
    median = df['Teaching Score'].median()
    
    # Fill null values with the calculated median of teaching score
    df['Teaching Score'].fillna(median, inplace=True)

```

```python
#Filling missing value of column citations score
if df['Citations Score'].isnull().any():
    # Calculate the median of the column citations score
    median = df['Citations Score'].median()
    
    # Fill null values with the calculated median of citations score
    df['Citations Score'].fillna(median, inplace=True)

```

```python
#Filling missing value of column Industry Income score
if df['Industry Income Score'].isnull().any():
    # Calculate the median of the column Industry Income score
    median = df['Industry Income Score'].median()
    
    # Fill null values with the calculated median of Industry Income score
    df['Industry Income Score'].fillna(median, inplace=True)

```

```python
#Filling missing value of column International Outlook score
if df['International Outlook Score'].isnull().any():
    # Calculate the median of the column International Outlook score
    median = df['International Outlook Score'].median()
    
    # Fill null values with the calculated median of International Outlook score
    df['International Outlook Score'].fillna(median, inplace=True)

df.isnull().sum()
```

```python
#Taking Teaching Score', 'Research Score', 'Citations Score','Industry Income Score','International Outlook Score these all columns into a column
numerical_columns = ['Teaching Score', 'Research Score', 'Citations Score','Industry Income Score','International Outlook Score']

# Calculate overall score by averaging scores from selected numerical columns
df['Overall Score'] = df[numerical_columns].mean(axis=1)
```

```python
#Dropping old overall score column permanently
df.drop(['OverAll Score'],axis=1,inplace=True)
```

```python
df = df.sort_values(by='Overall Score', ascending=False)

# Creating a new 'New_Rank' column based on the sorted order
df['University Ranking'] = range(1, len(df) + 1)

print("DataFrame with new rank based on 'Overall_Score':")

df
# Move the 13th column to the 1st position
df.insert(0, 'University Ranking', df.pop('University Ranking'))

df
```

```python
#Removing comma from no of student column to make values int type
df['No of student'] = df['No of student'].str.replace(',', '').astype(int)
df
```

```python
#Checking unique value in column international student
df['International Student'].unique()
```

```python
#Removing only % value into null
#df['International Student'] = df['International Student'].str.replace('%', '', regex=True)
#Changing percentage value into float by diving it by 100
df['International Student'] = pd.to_numeric(df['International Student'], errors='coerce') / 100
```

```python
df
```

```python
#Dropping the false elements present in clumn Female:AMle ratio
df.dropna(subset=['Female:Male Ratio'],inplace=True)
```

```python
# Making two column for female and male separately
split_ratio = df['Female:Male Ratio'].str.split(':', expand=True).astype(int)

# Calculating female and male number based on the ratio and total number of students
df['Female'] = (split_ratio[0] / (split_ratio[0] + split_ratio[1])) * df['No of student']
df['Male'] = (split_ratio[1] / (split_ratio[0] + split_ratio[1])) * df['No of student']

# Round the counts to integers
df[['Female', 'Male']] = df[['Female', 'Male']].round().astype(int)

# Drop the 'Ratio' column as we dont need this
df.drop(columns='Female:Male Ratio', inplace=True)

print("DataFrame with 'Female' and 'Male' columns based on Ratio and Total_Students:")
```

```python
# Move the Female column to the 4th position
df.insert(4, 'Female', df.pop('Female'))
```

```python
# Move the Male column to the 4th position
df.insert(5, 'Male', df.pop('Male'))
```

```python
# Make a Line Chart for variable citations score and international outlook score
plt.figure(figsize=(8, 5))
plt.plot(df['University Ranking'], df['Citations Score'], label='Citations Score', marker='o')
plt.plot(df['University Ranking'], df['International Outlook Score'], label='International Outlook Score', marker='o')
plt.title('International outlook score and Citations Score vs University Ranking')
plt.xlabel('University Ranking')
plt.ylabel('International outlook Score and Citations score')
plt.legend()
plt.tight_layout()
```

Variables plotted:
Blue= Citations score
Orange= International outlook score

Interpretion:
Here the two variables are ploted on the basis of increasing university ranking. Here we can observe the behaviour of two variables with increase in ranks and can interpret the result of relation between them.

Conclusion: 
The line chart is overall necessary for comparing the effect of two variable over a single base variable and observe thier affect with that third variable over the dataset.

***Bivariate Analysi***


```python
#Scatter plot between No of student per staff vs Teaching Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='No of student per staff', y='Teaching Score', data=df)

# Adding labels and title
plt.title('Bivariate Analysis: No of student per staff vs Teaching Score')
plt.xlabel('Number of  student per staff')
plt.ylabel('Teaching Score')

# Show the plot
plt.show()
```

Here X axis represent number of student per staff and y axis refers to teaching score and every single points in graph represent single university. This visualization helps in finding correlation among these two variables. Here we can see that the correlation between these two variable is highly negative. Here we can see as number of teaching student per staff increases the teaching score decrease. Here there is presence of a tendency for university with higher student to staff ratio in teaching to have less teaching score. But there are very few of outlier present that does not go with trend.

```python
columns_of_interest = ['No of student per staff', 'Teaching Score']
heatmap_data = df[columns_of_interest]

# Create a heatmap for the selected columns
# The annot=True parameter adds annotations with correlation values, and the cmap parameter sets the color map.
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')

# Adding title
plt.title('Heatmap: Correlation between Number of teaching student per staff and Teaching Score')

# Show the plot
plt.show()
```

From this figure also we can see that these two variables are moderately negatively correlated with each other. Here This might imply that in universities where there are fewer teaching staff members available for each student, quality of the teaching are measured through teaching score has meant to be lower. This show inverse type relation between these two variables.
